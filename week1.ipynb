```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
#list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/777"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image menjadi ukuran separuh
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # convert warna menjadi RGB
  final_frame = cv.hconcat((image, image_2))      # menyatukan gambar
  cv2_imshow(final_frame)                         # menampilkan gambar
  print('\n')
```
