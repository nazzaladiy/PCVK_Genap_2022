```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

```python
#list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/777"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          #read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) #resize image menjadi ukuran separuh
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # convert warna menjadi RGB
  final_frame = cv.hconcat((image, image_2))      # menyatukan gambar
  cv2_imshow(final_frame)                         # menampilkan gambar
  print('\n')
```

```python
# Melihat ukuran file image
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

```python
# mengakses pixel dengan memberi garis diagonal menyilang = Accessing pixel by creating lines in image

image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# membuat garis horizontal ditengah image
for y in  range (lebar):
  image_3[int((tinggi)/2),y] = [0,0,0]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

**TUGAS**

```python
# Garis vertikal
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in  range (tinggi):
  image_3[y, int((lebar)/2)] = [0,0,0]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)

# Garis diagonal 
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Garis Menyilang diagonal dari image keluaran
cv.line(image_3, (0,0), (lebar, tinggi), (0,0,0))

cv.line(image_3, (400,0), (0, tinggi), (0,0,0))

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
